```python
import requests
```

We use the `/works` api to list all works related to the orcid. This gives a summary of all works, so citation information is not included. We collect the `put-code` of all works to retrieve the citation information later.

```python
response = requests.get('https://pub.orcid.org/v3.0/{}/works'.format(orcid),
                        headers={"Accept": "application/orcid+json" })
record = response.json()
```

```python
record
```

```python
put_codes = []
for work in record['group']:
    put_code = work['work-summary'][0]['put-code']
    put_codes.append(put_code)
# put_code = put_codes[0]
```

```python
put_codes
```

We use the `/<orcid>/work/<put-code>` endpoint to retrieve the citation information for each record.

```python
citations = []
for put_code in put_codes:
    response = requests.get('https://pub.orcid.org/v3.0/{}/work/{}'.format(orcid, put_code), headers={"Accept": "application/orcid+json" })
    work = response.json()
    print(response.json())
    # print(work['citation'])
    # print(work['citation-value'])
    # citations.append(work)
    if work['citation'] is not None:
        # print(work['citation']['citation-value'])
        citations.append(work['citation']['citation-value'])
```

```python
import requests

citations = []

for put_code in put_codes:
    response = requests.get(
        f'https://pub.orcid.org/v3.0/{orcid}/work/{put_code}',
        headers={"Accept": "application/orcid+json"}
    )

    work = response.json()

    title = work['title']['title']['value']

    journal = None
    if work.get('journal-title'):
        journal = work['journal-title'].get('value')

    year = None
    if work.get('publication-date'):
        year = work['publication-date'].get('year', {}).get('value')

    doi = None
    external_ids = work.get('external-ids', {}).get('external-id', [])
    for ext in external_ids:
        if ext.get('external-id-type') == 'doi':
            doi = ext.get('external-id-value')
            break

    citation = {
        'title': title,
        'journal': journal,
        'year': year,
        'doi': doi
    }

    citations.append(citation)

print(citations)
```

```python
with open('output.bib', 'w') as bibfile:
    for citation in citations:
        bibfile.write(str(citation))
        bibfile.write('\n')
```
